Convert video file to an audio file

```python
import moviepy.editor as mp
import torchaudio
from transformers import WhisperProcessor, WhisperForConditionalGeneration, pipeline


# Insert Local Video File Path
clip = mp.VideoFileClip("./project_short_video/sample_video_lenny_1hour.mp4")

# Insert Local Audio File Path
clip.audio.write_audiofile(r"tmp_long_audio.wav")
```

```python
waveform, sample_rate = torchaudio.load("tmp_long_audio.wav")
resampler = torchaudio.transforms.Resample(sample_rate, 16000)
waveform = resampler(waveform)[0]
```

```python
import whisper
model = whisper.load_model("tiny")
```

```python
results = model.transcribe("./tmp_long_audio.wav", word_timestamps = True)
```

```python
print(results["segments"][0])
```

```python
for r in results["segments"]:
    start = str(int(r["start"]))
    end = str(int(r["end"])) 
    print(  start + ' - ' + end + ' ' + r['text'] )
```
